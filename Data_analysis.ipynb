```python
# After viewing the multiple columns in the listings.csv from the data_description.txt, 
# the following variables were picked for further analysis and dropped variables like date_scraped etc.
list_df = listings_df[['id','name','summary','longitude','latitude','space','description','instant_bookable','neighborhood_overview','neighbourhood_cleansed','host_id','host_name','host_since',
                 'host_response_time','street', 'zipcode','review_scores_rating','property_type','room_type','accommodates','bathrooms','bedrooms','beds','reviews_per_month','amenities','cancellation_policy','number_of_reviews','price']]
list_df.head()
```

```python
# Replace NaN values with 0
list_df.fillna(0, inplace=True)

# Extract prices from list_df into price_df
price_df = list_df['price']
# Create an empty prices list
prices=[]

# Convert prices from list_df into float values and append it in prices list
for p in price_df:
    p = float(p[1:].replace(',',''))
    prices.append(p)
    
# Replace the price column in the original list_df with the new prices
list_df['price'] = prices
```

```python
# Remove listings with 0 for  bedrooms, bathrooms, accomodates, price, beds, review_scores_rating, reviews_per_month
conditions = (
    (list_df['bedrooms'] > 0) &
    (list_df['bathrooms'] > 0) &
    (list_df['accommodates'] > 0) &
    (list_df['price'] > 0) &
    (list_df['beds'] > 0) &
    (list_df['review_scores_rating'] > 0) &
    (list_df['reviews_per_month'] > 0)
)
list_df = list_df[conditions]

list_df.head()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Downloading the VADER lexicon for sentiment analysis 
nltk.download('vader_lexicon')

# Initializing the VADER sentiment intensity analyzer
sid = SentimentIntensityAnalyzer()

# Filling missing comments with an empty string
reviews_df['comments'] = reviews_df['comments'].fillna('')

# Calculating sentiment scores for each review
reviews_df['sentiment_score'] = reviews_df['comments'].apply(lambda x: sid.polarity_scores(x)['compound'])
```

```python
# Merge 'list_df' and 'reviews_df' DataFrames based on 'listing_id'
merged_lr_df = pd.merge(list_df, reviews_df, left_on='id', right_on='listing_id', how='inner')

# Plotting
plt.figure(figsize=(10, 6))
sb.boxplot(x='room_type', y='sentiment_score', data=merged_lr_df, palette='Set1')
plt.title('Review Sentiment by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Sentiment Score')
plt.show()
```

#### Amenities Comparison

```python

# Example of amenities data in a single column
amenities_column = list_df['amenities']

# Function to extract amenities from the string and return a list
def extract_amenities(amenities_string):
    # Remove curly braces and quotes
    amenities_string = amenities_string.replace('{', '').replace('}', '').replace('"', '')
    # Split the string into a list of amenities
    amenities_list = amenities_string.split(',')
    return amenities_list

# Apply the function to extract amenities for each listing
list_df['amenities_list'] = list_df['amenities'].apply(extract_amenities)

# Create a list of all unique amenities
all_amenities = [amenity.strip() for amenities_list in list_df['amenities_list'] for amenity in amenities_list]
unique_amenities = list(set(all_amenities))

# Count the frequency of each amenity
amenities_freq = {amenity: all_amenities.count(amenity) for amenity in unique_amenities}

# Convert the dictionary to a DataFrame for plotting
amenities_df = pd.DataFrame.from_dict(amenities_freq, orient='index', columns=['Frequency'])
amenities_df = amenities_df.sort_values(by='Frequency', ascending=False)

# Plotting
plt.figure(figsize=(20, 10))
amenities_df.head(10).plot(kind='bar', color='lightgreen')
plt.title('Top Amenities by Frequency')
plt.xlabel('Amenity')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

####  Seasonal Trends

```python
# Convert 'date' column to datetime format
calendar_df['date'] = pd.to_datetime(calendar_df['date'])

# Merge 'calendar_df' with 'list_df' based on 'listing_id'
merged_df = pd.merge(calendar_df, list_df[['id', 'room_type']], left_on='listing_id', right_on='id', how='inner')

# Extract month from the 'date' column
merged_df['booking_month'] = merged_df['date'].dt.month

# Filter out unavailable listings ('available' column equals 't')
available_bookings = merged_df[merged_df['available'] == 't']

# Group by month and room type and count the number of bookings
seasonal_data = available_bookings.groupby(['booking_month', 'room_type']).size().reset_index(name='number_of_bookings')

# Plotting
plt.figure(figsize=(10, 6))
sb.lineplot(x='booking_month', y='number_of_bookings', hue='room_type', data=seasonal_data, marker='o')
plt.title('Seasonal Trends in Booking by Room Type')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(range(1, 13))
plt.grid(True)
plt.legend(title='Room Type')
plt.show()
```

### Analyzing if any particular ammenity results in higher prices.

```python
import nltk
from nltk.corpus import stopwords
import re

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Create a dataframe of the words that appear in the ammenities section of the most expensive listings

ami_df = list_df[['amenities','price','id',]]
ami_df_topper = ami_df.sort_values('price',ascending=[0])
ami_df_top =ami_df_topper.head(30)
allemenities = ''
for index,row in ami_df_top.iterrows():
    p = re.sub('[^a-zA-Z]+',' ', row['amenities'])
    allemenities+=p

allemenities_data=nltk.word_tokenize(allemenities)
filtered_data=[word for word in allemenities_data if word not in stopwords.words('english')] 
wnl = nltk.WordNetLemmatizer() 
allemenities_data=[wnl.lemmatize(data) for data in filtered_data]
allemenities_words=' '.join(allemenities_data)
```

```python
from wordcloud import WordCloud, STOPWORDS

wordcloud = WordCloud(width = 1000, height = 700, background_color="white").generate(allemenities_words)
plt.figure(figsize=(15,15))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

From the above wordcloud, we can see that the listings with the highest prices have ammenities such as 'Washer', Dryer', 'Heating', 'Wireless Internet', 'Smoke Detector', 'Free Parking', 'Kid'.
